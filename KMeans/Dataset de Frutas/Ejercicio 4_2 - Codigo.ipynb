```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import KMeans
```

```python
df = pd.read_excel('Date_Fruit_Datasets.xlsx', 'Date_Fruit_Datasets')
```

```python
df.head()
```

```python
values = df.iloc[:, -1].unique()
values
```

```python
df['Class'] = df['Class'].replace((values), ([i for i in range(len(values))]))
df.head()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
scaler = StandardScaler()
X = np.array(scaler.fit_transform(X))
df = np.c_[X, y]
df = pd.DataFrame(df)
df.head()
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=51)
```

```python
pca = PCA(2)
data = np.c_[X_train, y_train]
data = pca.fit_transform(data)
```

```python
kmeans = KMeans.KMeans(7, data, 1500)
label, centroids = kmeans.fit()
plt.axis([-8, 13, -6.5, 7])
kmeans.plot(label, data, centroids)
```

```python
data_test = np.c_[X_test, y_test]
data_test = pca.fit_transform(data_test)
```

```python
label_test = kmeans.predict(data_test, centroids)
```

```python
plt.axis([-8, 13, -6.5, 7])
kmeans.plot(label_test, data_test, centroids)
```
